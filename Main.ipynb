```python
# Import Requisite Libraries
import sqlite3
import numpy as np
import plotly as px
import pandas as pd
import sklearn as sk
```

```python
# Connect to Database
con = sqlite3.connect("Database.db")
cur = con.cursor()

# Read Data (Credit to Ai0, https://onedrive.live.com/view.aspx?resid=15C4097138751CAF%21721339&authkey=!AHKvMDy7MhWxN7g)
from pandas.api.types import CategoricalDtype
data = pd.read_excel("Cutoffs.xlsx", usecols = "B:K", names = ["Name", "Cutoffs", "Group", "Focus", "No", "Duration", "Cheerful Carnival", "Max Bonus", "Max Bonus (No Gacha)", "Max Bonus (No 4 Stars)"], dtype = {"Group": CategoricalDtype(), "Focus": CategoricalDtype(), "Cheerful Carnival": CategoricalDtype()}, sheet_name = None)

# Create Table for each Sheet
for tier, values in data.items():
    values.to_sql(tier, con, if_exists = "replace")
```
